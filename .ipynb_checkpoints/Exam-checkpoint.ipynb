```python
# Задача 1 - прочитать n строк и вывести m самых часто встречаемых слов и их количество

import re

n, m = map(int, input().split())

lines = ''
for i in range(n):
    lines += input().strip()
    lines += ' '

pattern = r'\b(\w+)\b'
words = re.findall(pattern, lines)
#print(words)
uniqueWords = list(set(words))
wc = []
for word in uniqueWords:
    wc.append([word, words.count(word)])
    
wc.sort(key = lambda x: x[1], reverse = True)

for j in range(m):
    print(wc[j][0], wc[j][1])
```

```python
# Задача 2 - считать матрицу до символа "===", вывести матрицу того же размера,
#у которой единицы находятся в клетках, у которых среди соседей не меньше 3 и не больше 6 единиц
#В случае одной строки/столбца элемент сам себе сосед по этому направлению

task_matrix = []
while True:
    line = input().strip()
    if line == '===':
        break
    task_matrix.append([int(i) for i in line.split()])
print(task_matrix)

for i in range(len(task_matrix)) : #сначала идем по строкам
    for j in range(len(task_matrix[i])) : #теперь идем по столбцам
        if (i == len(task_matrix) - 1) : #проверяем, находится ли элемент в крайней строке
            ai = 0 #если да, заменяем индекс на 0
        else:
            ai = i + 1 #если нет, просто прибавляем к индексу единицу
        if (j == len(task_matrix[i]) - 1) :
            aj = 0
        else:
            aj = j + 1
        
        # в этой строке считаем сумму всех элементов по всем сторонам от выбранной точки
        k = task_matrix[i-1][j-1] + task_matrix[i][j-1] + task_matrix[ai][j-1] + task_matrix[i-1][j] + task_matrix[ai][j] + task_matrix[i-1][aj] + task_matrix[i][aj] + task_matrix[ai][aj]
        if 3 <= k <= 6:
            print(1, end = ' ')
        else:
            print(0, end = ' ')
        if (j == len(task_matrix[i]) - 1):
            print()
```

```python
# Задача 3 - написать класс, который можно создать либо из числа, либо из двух и более объектов класса
#Класс имеет метод, который возвращает либо обратное число (если в классе число), либо сумму первых минус последнее

class Difference:
    def __init__(self, *args):
        if len(args) == 1:
            #print('One argument')
            if isinstance(args[0], (int, float)):
                #print('The only argument is a number')
                self.val = args[0]
                self.fromnumber = True
            else:
                self.val = 0
        else:
            self.val = []
            self.fromnumber = False
            self.args = args
            for arg in args:
                if isinstance(arg, Difference):
                    if isinstance(arg.val, (int, float)):
                        self.val.append(arg.val)
                    elif isinstance(arg.val, list):
                        self.val.extend(arg.val)
                
    def evaluate(self):
        if self.fromnumber:
            return -self.val
        else:
            return(sum(map(lambda x: x.evaluate(), self.args[:len(self.args)-1])) - self.args[len(self.args)-1].evaluate())
        
a = Difference(5)
print(a.evaluate()) # -5
b = Difference(12)
c = Difference(a, b)
d = Difference(a, b, c)
print(c.evaluate()) # 7
print(d.evaluate()) # -24
print(a.evaluate()) # -5
```

```python
# Задача 4 - реализовать генератор для временной шкалы

# Все не так - необходимо реализовать добавление величины с временем добавления и временем удаления
#а не просто добавлять и удалять
# Структурой выберем список словарей, где ключ - значение, а значение по ключу - список списков [время_доб, время_уд]
# Необходимо понять, что происходит в ситуации: добавлено в t=2, добавлено в t=5, удалено в t=6, удалено в t=9

import itertools

class Numbers:
    def __init__(self):
        self.values = {}
    
    def add(self, x, t):
        # добавить число
        if x not in self.values.keys():
            #создали список списков
            self.values[x] = [[]]
            # добавили момент рождения
            self.values[x][0].extend([t, 0])
        else:
            self.values[x].append([t, 0])

    def remove(self, x, t):
        # удалить число
        # если число добавлено
        #if x in self.values.keys():
            #for 
        self.values[x][1] = t
        
    def query(self, t):
        # вернуть объект-генератор – результат запроса
        for key in self.values.keys():
            if key <= t:
                yield self.values[key]
                
sample = Numbers()
sample.add(1, 1)
sample.add(2, 10)

print(sample.values)

print(15)
for i in sample.query(15):
    print(i) # 1, 2
print()

print(6)
for i in sample.query(6):
    print(i) # 1
print()
# в момент 6 мы не увидим 2, поскольку оно было добавленно позднее

sample.remove(2, 8)

print(14)
for i in sample.query(14):
    print(i)# 1, 2
print()
# мы увидим 2, так как оно было добавлено в момент 10, а удалено в момент 8

sample.remove(1, 3)
sample.add(1, 5)
print(2)
for i in sample.query(2):
    print(i)# 1
print()
# 1 было добавлено, но еще не удалено

print(4)
for i in sample.query(4):
    print(i)# пустая последовательность
print()
# 1 было добавлено и удалено

print(6)
for i in sample.query(6):
    print(i)
print()
```

```python
# Задача 7 - проходимость уровня ниндзей
# Дано 3 строки: 1 - высота_уровня высота_прыжка, 2 - левая стенка, 3 - правая стенка

# Применяем решение при помощи графа
# Граф - это словарь со структурой {parent:[children]}

# Если просто собрать такой граф, то попытка найти путь обернется зацикливанием
# Чтобы избавиться от зацикливания, можно удалять из графа все посещенные точки. Это сложно, поскольку требует рекурсивного создания копий.
# Кроме того, можно вызывать None, если текущий узел уже есть в пути.

'''
import sys

with open('Exam_task6_1.txt') as inf:
    n, k = map(int, inf.readline().split())
    leftWall = inf.readline().strip()
    rightWall = inf.readline().strip()
'''
    
    
n, k = map(int, input().split())
leftWall = input().strip()
rightWall = input().strip()


'''
print(n,k)
print(leftWall)
print(rightWall)
'''

def find_path(graph, parent, child, path=None):
    if path == None:
        path = []
    # Попытка бороться с зацикливанием - вызываем None и закрываем ветку, которая привела к тому же узлу
    if parent in path:
        return None
    path.append(parent)    
    if parent == child:
        return path
    if parent not in graph.keys():
        return None
    for node in graph[parent]:
        newpath = find_path(graph, node, child, path)
        if newpath:
            return newpath
    return None



# Строим граф путей из каждого участка стены, на котором можно стоять

ways = {}
# для начала строим граф для leftWall
# для того, чтобы работать в 0-массивах, занимаем в нашем коде n на n-1
for i in range(n):
    # если на участке стены можно стоять - добавляем его в граф в качестве родителя
    if leftWall[i] != 'X':
        ways['l'+str(i)] = []
        
        # проверяем, можно ли сделать шаг вниз
        # если это не самый нижний участок на стене,
        if i != 0:
            # и участке ниже можно стоять,
            if leftWall[i-1] != 'X':
                # добавляем участок ниже в дети текущего участка
                ways['l'+str(i)].append('l'+str(i-1))
        
        # проверяем, можно ли сделать шаг вверх
        # если это последний участок стены, значит можно за один шаг вверх пройти уровень
        if i == n-1:
            # добавляем в дети этого участка специальный символ "выхода"
            ways['l'+str(i)].append('The end')
        # иначе, если на участке на шаг выше вверх можно стоять
        elif leftWall[i+1] != 'X':
            # добавляем участок выше в дети текущего участка
            ways['l'+str(i)].append('l'+str(i+1))
        
        # проверяем, можно ли прыгнуть на противоположную стенку
        # если мы за один прыжок выпрыгиваем из уровня
        if i+k > n-1:
            # добавляем в дети этого участка специальный символ "выхода" 
            ways['l'+str(i)].append('The end')
        # иначе, если при прыжке на противоположную стену на участке можно стоять,
        elif rightWall[i+k] != 'X':
            # добавляем участок после прыжка в дети текущего участка. Не забываем поменять букву стены!
            ways['l'+str(i)].append('r'+str(i+k))
            
# Теперь строим граф для rightWall
for i in range(n):
    # если на участке стены можно стоять - добавляем его в граф в качестве родителя
    if rightWall[i] != 'X':
        ways['r'+str(i)] = []
        
        # проверяем, можно ли сделать шаг вниз
        # если это не самый нижний участок на стене,
        if i != 0:
            # и участке ниже можно стоять,
            if rightWall[i-1] != 'X':
                # добавляем участок ниже в дети текущего участка
                ways['r'+str(i)].append('r'+str(i-1))
        
        # проверяем, можно ли сделать шаг вверх
        # если это последний участок стены, значит можно за один шаг вверх пройти уровень
        if i == n-1:
            # добавляем в дети этого участка специальный символ "выхода"
            ways['r'+str(i)].append('The end')
        # иначе, если на участке на шаг выше вверх можно стоять
        elif rightWall[i+1] != 'X':
            # добавляем участок выше в дети текущего участка
            ways['r'+str(i)].append('r'+str(i+1))
        
        # проверяем, можно ли прыгнуть на противоположную стенку
        # если мы за один прыжок выпрыгиваем из уровня
        if i+k > n-1:
            # добавляем в дети этого участка специальный символ "выхода" 
            ways['r'+str(i)].append('The end')
        # иначе, если при прыжке на противоположную стену на участке можно стоять,
        elif leftWall[i+k] != 'X':
            # добавляем участок после прыжка в дети текущего участка. Не забываем поменять букву стены!
            ways['r'+str(i)].append('l'+str(i+k))
            
# Теперь делаем проверку - существует ли в графе путь от ключа "l0" до ключа "The end"
#print(ways)

if find_path(ways, 'l0', 'The end'):
    print('YES')
else:
    print('NO')
```

```python
# Задача 8 - работаем с файловой структурой и файлами json
# Подходящий файл - .json с JSON-структурой внутри
# Интересный файл - есть поля kind и age в любом регистре
# Хороший файл - значение kind равно human, значение age >= 18

import os, os.path, re, json

good = 0

for current_dir, dirs, files in os.walk('./data/'):
    for file in files:
        pat_json = r'\.json'
        if re.search(pat_json, file):
            with open(current_dir + '/' + file) as inf:
                text = inf.read().lower()
                infile = json.loads(text)
                allKeys = infile.keys()
                if 'kind' in allKeys and 'age' in allKeys:
                    if infile['kind'].lower() == 'human' and infile['age'] >= 18:
                        good += 1
                        
print(good)
```

```python
# Задача 2 - обход папки

print(os.getcwd())

l = []

for current_dir, dirs, files in os.walk("./main/"):
    if list(filter(lambda x: x.endswith('.py'), files)) != []:
        l.append(current_dir[2:])

l.sort()

with open('24_task2_ans.txt', 'w') as ouf:
    for d in l:
        ouf.write(d + '\n')
        print(d)
```
